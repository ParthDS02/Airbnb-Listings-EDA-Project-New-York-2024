```python
# price dependency on neighbourhood
sns.barplot(data=df, x='neighbourhood_group', y='price', hue='room_type')
plt.show()
```

```python
# Bar plot of price dependency on neighbourhood with annotations
plt.figure(figsize=(10, 6))
barplot = sns.barplot(data=df, x='neighbourhood_group', y='price', hue='room_type')

# Add annotations
for p in barplot.patches:
    # Get the height of the bar (y-value) and its x-coordinate
    height = p.get_height()
    barplot.annotate(f'{height:.2f}',  # Format annotation with 2 decimal places
                     (p.get_x() + p.get_width() / 2., height),  # Position the annotation
                     ha='center', va='center',  # Horizontal and vertical alignment
                     xytext=(0, 5),  # Offset the text slightly above the bar
                     textcoords='offset points')

# Set titles and labels
plt.title('Price Dependency on Neighbourhood Group by Room Type')
plt.ylabel('Average Price')
plt.xlabel('Neighbourhood Group')
plt.show()
```

```python
# number of reviews and price rel
plt.figure(figsize=(10, 10))
plt.title("Locality and Review Dependency")
sns.scatterplot(data=df, x='number_of_reviews', y='price', hue='neighbourhood_group')
plt.show()
```

This scatter plot shows the relationship between the price of Airbnb listings and the number of reviews they have received, segmented by neighborhood groups: Brooklyn (blue), Manhattan (orange), Queens (green), Bronx (red), and Staten Island (purple).

### Insights from this visualization:

1. Price Distribution: Listings in Manhattan (orange) tend to have the highest prices, with some exceeding $1,000. Brooklyn (blue) follows with a wide range of prices but with fewer high-end listings compared to Manhattan.

2. Review Count: Many listings have fewer than 250 reviews, with some listings, particularly in Manhattan, having fewer reviews despite high prices. This could suggest that luxury or exclusive properties are not reviewed as frequently.

3. Popular Listings: Listings with higher numbers of reviews (500+), especially in Brooklyn and Queens, seem to have more moderate prices, suggesting higher turnover and popularity.

4. Outliers: Some listings in Manhattan have very high prices but relatively low review counts, possibly indicating newer or more exclusive properties.

This is a multi-variable plot comparing different features of Airbnb listings: price, minimum nights, number of reviews, and availability over the past 365 days.

### Insights from this visualization:

1. Price: Listings are predominantly concentrated in lower price ranges, with most priced below $300. However, there are a few high-priced outliers (above $1,000).

2. Minimum Nights: Most listings have short minimum stay requirements (less than 10 nights), with few exceptions for longer stays.

3. Number of Reviews: Listings with fewer reviews dominate, with occasional high-review listings, possibly indicating popular or long-standing hosts.

4. Availability: There is a clear clustering of availability (365-day availability), suggesting that many listings have maximum availability throughout the year.

### Correlations:
-- Higher-priced listings tend to have fewer reviews, which could indicate luxury properties with less turnover.

-- Availability doesn't show a strong correlation with price or reviews, suggesting that various types of listings (in terms of price and review counts) may be equally available throughout the year.

```python
# heat map - correlation of one variable with others for numerical column

corr = df[['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365', 'beds']].corr()
corr

plt.figure(figsize=(8, 6))
sns.heatmap(data=corr, annot=True)
plt.show()
```

This heatmap represents the correlation matrix for various features related to Airbnb listings, showing how different variables relate to each other in terms of linear correlation coefficients (ranging from -1 to 1). Here’s a breakdown of the analysis based on the correlations presented

### Analysis of the Correlation Matrix:

1. Price Correlations:  Number of Reviews (0.63): There is a strong positive correlation between number_of_reviews and price. This suggests that listings with more reviews tend to have higher prices, possibly indicating that more popular listings are more expensive.
Reviews per Month (0.63): Similarly, reviews_per_month also shows a strong positive correlation with price. Listings that receive more reviews each month are likely priced higher, indicating a relationship between activity level and pricing.
Beds Correlation:

2. Price (0.42):   The number of beds has a moderate positive correlation with price. This indicates that listings with more beds generally command higher prices, which is typical in the market for larger accommodations.
Availability Correlation:

3. Reviews per Month (-0.04):   There is a very weak negative correlation between availability_365 and reviews_per_month. This might suggest that listings available all year round may not necessarily receive more reviews, potentially indicating that a balance between availability and guest engagement is crucial for receiving more reviews.
Geographical Influence:

4. Latitude and Longitude:   The correlations with latitude and longitude are quite weak, indicating that geographical location does not strongly influence the price or number of reviews. This could imply that factors other than location are more significant in determining pricing.
Minimum Nights:

5. Weak Correlation with Other Variables:   The minimum_nights variable shows weak correlations with other features, suggesting that the minimum stay requirement does not significantly affect price or the number of reviews.

### Conclusion 

Through analyzing the Airbnb data, a few clear trends emerge. Listings in Manhattan tend to be priced higher but have fewer reviews, likely due to their premium nature. On the other hand, more affordable listings in areas like Brooklyn and Queens generate significantly more guest interaction, with higher numbers of reviews, suggesting that affordability drives engagement.

Room type plays a significant role too—entire homes are the most expensive and widespread, while private rooms are more dispersed and generally lower-priced. Interestingly, while the geographical location (latitude and longitude) doesn't seem to strongly influence price or reviews, the number of reviews and activity level (reviews per month) are key indicators of pricing strategy.

Additionally, being available year-round doesn't guarantee more reviews, implying that guest engagement and targeted marketing may matter more than sheer availability.

Thus, successful Airbnb hosts should focus on offering competitively priced listings, optimizing guest experiences to drive reviews, and carefully considering room type and guest interaction, rather than relying solely on location or constant availability. This balanced approach is crucial in a competitive market like New York.
